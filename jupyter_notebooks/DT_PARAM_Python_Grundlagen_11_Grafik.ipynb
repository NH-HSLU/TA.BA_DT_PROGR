```python
xs = []
ys = []

name = "Anna"
gender = "F"
state = "AK"

with open("./Data/names.csv","r") as file:
    counter = 0
    
    for line in file:
        
        counter = counter + 1
        data = line.strip().split(",")
        if data[1] == name and data[3] == gender and data[4] == state:
            xs.append(int(data[2]))
            ys.append(int(data[5]))
            
%matplotlib inline
import matplotlib.pyplot as plt


plt.plot(xs,ys)

plt.xlabel('Jahre')
plt.ylabel('Anzahl')

plt.show()
```
